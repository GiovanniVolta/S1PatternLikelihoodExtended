```python
plt.figure(figsize=(15,8))
# Attenuazione ad 1 MeV
I0 = 1e4 # N
mu_over_rho_ = 1e-1 # cm^2/g

def att(x, I0=1e4, rho = 2.86, mu_over_rho= 1e-1):
    return I0*np.exp(-x/(mu_over_rho*rho))
z = np.linspace(0, 96, 1000)

I = att(z)

plt.plot(z-96, I/I0, 'r--')
plt.xlabel('z [cm]', fontsize=26)
plt.ylabel('Attenuation @ 1 MeV', fontsize=26)
plt.tick_params(labelsize=20)
#plt.xscale('log')
plt.xlim(-97,-70)
```
